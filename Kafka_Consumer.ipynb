```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
```

```python
consumer = KafkaConsumer(
    'demo_testing2',
    bootstrap_servers=['3.142.136.127:9092'], #add your IP here
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
for c in consumer:
   print(c.value)
```

```python
s3 = S3FileSystem()
```

```python
for count, i in enumerate(consumer):
    with s3.open("s3://kafka-stockmarket-project-tanmaykapure/stock_market_{}.json".format(count), 'w') as file:
        json.dump(i.value, file)    
```
