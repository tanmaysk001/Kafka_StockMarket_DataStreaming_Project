```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
import boto3
```

```python
consumer = KafkaConsumer(
    'demo_testing2',
    bootstrap_servers=['3.147.57.49:9092'], #add your IP here
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
for c in consumer:
   print(c.value)
```

```python
s3 = S3FileSystem()
```

```python
s3_access_point_arn = 'arn:aws:s3:us-east-2:your_account_id:accesspoint/your_access_point_name'

# Write messages to S3 using the access point
for count, i in enumerate(consumer):
    s3.put_object(
        Bucket=s3_access_point_arn,
        Key='stock_market_{}.json'.format(count),
        Body=json.dumps(i.value)
    )
```
